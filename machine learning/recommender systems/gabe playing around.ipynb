```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
columns = ['user_id','item_id','rating','timestamp']
df = pd.read_csv('u.data',sep='\t',names = columns)
```

```python
df.head()
```

```python
movie_titles = pd.read_csv('Movie_Id_Titles')
```

```python
df = pd.merge(df,movie_titles,on='item_id')
```

```python
df.head()
```

```python
# group the movies by the highest rating
df.groupby('title')['rating'].mean().sort_values(ascending=False)
```

```python
# group the movies by the most number of reviews
df.groupby('title')['rating'].count().sort_values(ascending=False)
```

```python
# now make a new dataframe with the means and counts of ratings
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
```

```python
ratings.head()
```

```python
ratings['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
```

```python
ratings.head()
```

```python
# look at a couple of plots

ratings['rating'].hist(bins=70)
```

```python
ratings['num of ratings'].hist(bins=70)
```

```python
# now create the pivot table to visualize
# user id ratings (y axis) by movie title (x axis)

moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
```

```python
moviemat.head()
```

```python
df['title'][df.title.str.startswith('B')].unique()
```

```python
adams_family_ratings = moviemat['Addams Family Values (1993)']
```

```python
moviemat.corrwith(adams_family_ratings)
similar_to_adams_family = moviemat.corrwith(adams_family_ratings)
```

```python
corr_adams_family = pd.DataFrame(similar_to_adams_family,
                                columns=['Correlation'])
corr_adams_family.dropna(inplace=True)
```

```python
corr_adams_family.head()
```

```python
corr_adams_family = corr_adams_family.join(ratings['num of ratings'])
```

```python
corr_adams_family[corr_adams_family['num of ratings']>100].sort_values('Correlation',ascending=False).head(10)
```

```python
sphere_ratings = moviemat['Sphere (1998)']
```

```python
similar_to_sphere = moviemat.corrwith(sphere_ratings)
```

```python
similar_to_sphere.head()
```

```python
corr_sphere = pd.DataFrame(similar_to_sphere,columns=['Correlation'])
corr_sphere.dropna(inplace=True)
```

```python
corr_sphere.head()
```

```python
corr_sphere = corr_sphere.join(ratings['num of ratings'])
```

```python
corr_sphere.sort_values('Correlation',ascending=False).head(15)
```

```python
corr_sphere[corr_sphere['num of ratings']<40].sort_values('Correlation',ascending=False).head(20)
```
