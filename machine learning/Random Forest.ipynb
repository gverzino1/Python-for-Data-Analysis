```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('kyphosis.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
sns.pairplot(df,hue='Kyphosis')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Kyphosis',axis=1)
y = df['Kyphosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
from sklearn import metrics
```

```python
# can see we're mislabeling about 10 cases from the test set
print(confusion_matrix(y_test,predictions))
print('\n \n')
print(classification_report(y_test,predictions))
print('\n \n')
print("Accuracy:",metrics.accuracy_score(y_test,predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# now apply random forests to the training sets with 200 estimators 
rfc = RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(X_train,y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test,rfc_pred))
print('\n \n')
print(classification_report(y_test,rfc_pred))
print('\n \n')
print("Accuracy:",metrics.accuracy_score(y_test,rfc_pred))
```

```python
from graphviz import Source
from sklearn import tree

# gives names to the outcomes classes
class_names = dtree.classes_
class_names

features = list(df.columns[1:])
features
```

```python
# to display the decision tree
Source(tree.export_graphviz(dtree,
                           out_file=None,
                            filled=True, 
                            rounded=True,
                            special_characters=True,
                            class_names=class_names,
                           feature_names=features))

# to save png and view it
#graph = Source(tree.export_graphviz(dtree,out_file=None,feature_names=features))
#graph.format = 'png'
#graph.render('dtree_render',view=True)
```

```python
# now lets attempt to improve this tree
# Gini is the probability of a random sample being classified incorrectly
new = DecisionTreeClassifier(criterion="gini",max_depth=3)
new = new.fit(X_train,y_train)
y_pred = new.predict(X_test)
print(classification_report(y_test,rfc_pred))
print('\n \n')
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```
