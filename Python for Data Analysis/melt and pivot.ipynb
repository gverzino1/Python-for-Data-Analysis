```python
data1 = pd.DataFrame({'Account':[123,321,324,234,675,454],
              'RA Date':['05/02/2020','05/02/2020','05/02/2020','05/03/2020','05/04/2020','05/02/2020'],
             'EOB Detail':[('7873 8920'),('0313'),('2322 0902'),('2321'),('6215 0932'),('8782 6215 0992')]})
```

```python
data1['EOB Detail'].str.split(expand=True)
```

```python
eobs = data1['EOB Detail'].str.split(expand=True)
```

```python
pd.concat([data1,eobs],axis=1)
```

```python
data1.join(eobs).replace(np.nan, '')
```

```python
data2 = data1.join(eobs)
```

```python
data2.rename(columns = {0:'EOB1',1:'EOB2',2:'EOB3'},inplace=True)
```

```python
pd.melt(data2,id_vars=['Account','RA Date'],value_vars=['EOB1','EOB2','EOB3']).drop('variable',axis=1).sort_values('Account')
```

```python
data3 = pd.melt(data2,id_vars=['Account','RA Date'],value_vars=['EOB1','EOB2','EOB3']).drop('variable',axis=1).sort_values('Account')
```

```python
data3.rename(columns={'value':'EOBs'},inplace=True)
```

```python
data3.reset_index(drop=True,inplace=True)
```

```python
data3['EOBs'].isna()
```

```python
data3.dropna(inplace=True)
```
