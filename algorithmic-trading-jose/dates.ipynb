```python
# to do time resampling you need datetime index
# this is a groupby for dates by the year (that's the A)
df.resample(rule='A')

# and now you can get a mean for everything by year
df.resample(rule='A').mean()

# do a mean for quarterly
df.resample(rule='Q').mean()
```

```python
df.resample(rule='A').max()
```

```python
# what if you just wanted to know the first data points for a time period
def first_day(entry):
    return entry[0]    # return data in the first index only
```

```python
# then apply your function
df.resample('A').apply(first_day)
```

```python
df['Close'].resample('M').mean().plot(kind='bar',figsize=(16,6))
```

```python
# note that is daily data
df.head()
```

```python
# to shift the data up one period, but you'll lose your last row of data too!
df.shift(periods=1).head()
```

```python
# you can also shift the index
# say you want all of january dailies to just be 01-31-2012 (for 2012)
df.tshift(freq='M').head()
```

```python
df['Open'].plot(figsize=(16,6))
```

```python
# Now you can smooth is out by using a rolling average (like SMA)
# use the 7 SMA, and check out the first 14 rows
df.rolling(7).mean().head(14)
```

```python
# to call open prices with an average of the close prices
# use a 50 SMA
df['Open'].plot()
df.rolling(window=50).mean()['Close'].plot(figsize=(16,6))
```

```python
# to make a legend byfirst assigning the MA to df['Close 30 Day MA'], 
# then referencing that in the plot along with the normal 'close' price
df['Close 30 Day MA'] = df['Close'].rolling(window=30).mean()
df[['Close 30 Day MA','Close']].plot(figsize=(16,6))
```

```python
# to get the average of all the points before it
# so the last point on y-axis is the average of all prices before
df['Close'].expanding().mean().plot(figsize=(16,6))
```

```python
# create bollinger bands
# Close 20 MA
df['Close: 20 Day Mean'] = df['Close'].rolling(20).mean()

# Upper Band = 20MA + 2*std(20)
df['Upper'] = df['Close: 20 Day Mean'] + 2*(df['Close'].rolling(20).std())

# Lower Band = 20MA - 2*std(20)
df['Lower'] = df['Close: 20 Day Mean'] - 2*(df['Close'].rolling(20).std())

# then plot the whole thing out
df[['Close','Close: 20 Day Mean','Upper','Lower']].tail(200).plot(figsize=(16,6))
```
